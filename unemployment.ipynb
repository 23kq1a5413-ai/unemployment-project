```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile

# CSV file path
csv_file_path = "/content/Unemployment_Rate_upto_11_2020.csv"

# Read CSV directly
df = pd.read_csv(csv_file_path)

# Remove missing values
df = df.dropna()

# =========================
# DATASET PREVIEW
# =========================
print("===== DATASET PREVIEW =====")
print(df.head())

# Convert Date column
df[' Date'] = pd.to_datetime(df[' Date'], dayfirst=True)

# =========================
# ANDHRA PRADESH DATA
# =========================
ap_df = df[df['Region'] == 'Andhra Pradesh']

print("\n===== ANDHRA PRADESH UNEMPLOYMENT RATE =====")
print(ap_df[[' Date', ' Estimated Unemployment Rate (%)']].head())

print("\n===== EMPLOYMENT IN ANDHRA PRADESH =====")
print(ap_df[[' Date', ' Estimated Employed']].head())
# =========================
# GRAPH 1
# Andhra Pradesh Unemployment Rate
# =========================
plt.figure(figsize=(12,5))
plt.plot(
    ap_df[' Date'],
    ap_df[' Estimated Unemployment Rate (%)'],
    marker='o'
)

plt.title("Andhra Pradesh Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

# =========================
# GRAPH 2
# Employment in Andhra Pradesh
# =========================
plt.figure(figsize=(12,5))
plt.bar(
    ap_df[' Date'],
    ap_df[' Estimated Employed']
)

plt.title("Employment in Andhra Pradesh")
plt.xlabel("Date")
plt.ylabel("Estimated Employed")
plt.xticks(rotation=45)
plt.show()

# =========================
# SUMMARY
# =========================
print("\n===== SUMMARY =====")
print("Average Unemployment Rate:",
      round(ap_df[' Estimated Unemployment Rate (%)'].mean(), 2), "%")

print("Maximum Unemployment Rate:",
      round(ap_df[' Estimated Unemployment Rate (%)'].max(), 2), "%")

print("Minimum Unemployment Rate:",
      round(ap_df[' Estimated Unemployment Rate (%)'].min(), 2), "%")

print("Average Employment:",
      round(ap_df[' Estimated Employed'].mean()))
```
